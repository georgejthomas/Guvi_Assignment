```python
import csv

def registration():
    # ACCEPTING EMAIL ID/USERNAME AND PERFORMING VERIFICATION
    while True:
        email = input("\nEnter NEW Email ID: ")

        if '@' in email and '.' in email and email.index('.') > (email.index('@') + 1):
            subs = email.split('@')

            if subs[0].isalpha() == True:
                #print("Email ID Accepted")
                break
            else:
                print("YOU can only have Alphabets before the @")
        else:
            print("Email ID must have special characters \'@\',\'.\' and \'.\' must not come immediately after \'@\'")

    # ACCEPTING PASSWORD AND PERFORMING VERIFICATION
    while True:
        print("")
        password = input("\nEnter NEW Password. \nPassword must be of 5 -16 characters long. \nPassword must have Minimum: \n1 UPPERCASE CHARACTER \n1 LOWERCASE CHARACTER \n1 DIGIT \n1 SPECIAL CHARACTER FROM [ @, $, !, %, _, #, (, ), &, * ]\nEnter here: ")

        special_char = ['@', '$', '!', '%', '_', '#', '(', ')', '&', '*']
        dig_cnt = spechar_cnt = lowchar_cnt = uppchar_cnt = 0

        if len(password) >= 5 and len(password) <= 16:
            for ch in password:
                if ch.isupper():
                    uppchar_cnt += 1
                elif ch.islower():
                    lowchar_cnt += 1
                elif ch.isdigit():
                    dig_cnt += 1
                elif ch in special_char:
                    spechar_cnt += 1

            if uppchar_cnt == 0 or lowchar_cnt == 0 or dig_cnt == 0 or spechar_cnt == 0:
                print("Password must have Minimum: \n1 UPPERCASE CHARACTER \n1 LOWERCASE CHARACTER \n1 DIGIT \n1 SPECIAL CHARACTER FROM [ @, $, !, %, _, #, (, ), &, * ]")
            else:
                #print("Password accepted")
                break
        else:
            print("Password Length should be between 5 to 16 characters")

    # ADDING CREDENTIALS TO FILE
    login_id_pswrd = [email, password]

    fname = "login_creds.csv"

    with open(fname, 'a') as f:
        csvWriter = csv.writer(f)

        if csvWriter.writerow(login_id_pswrd):
            print("Credentials added")
        else:
            print("Not Successful")

# --------------------------------------------------------------------------------

def login():
    attempt = 1
    chk = 0
    try:
        while attempt <= 2:
            email = input("\nEnter Email ID: ")
            password = input("\nEnter Password: ")
            
            fname = "login_creds.csv"
            
            with open(fname, 'r') as f:
                csvReader = csv.reader(f)
                
                for x in csvReader:
                    if email == x[0]:
                        if password == x[1]:  # If both email id & password are correct
                            print("Hello", x[0], "Login Successful")
                            return
                        elif password != x[1] and attempt < 2:  # If email id is correct but password is wrong
                            print("Incorrect Login Credentials. Try Again")
                            attempt += 1
                            chk = 2
                            break
                        else:
                            print("Incorrect Login Credentials. 2 ATTEMPTS OVER")
                            chk = 3
                            break
                    else:
                      chk = 0
                
                if chk == 0:
                    inval = input("Login credentials do NOT Exists \nWOULD YOU LIKE TO REGISTER NEW (Enter Y/y for YES or n/N for NO): ")
                    if inval in ('y', 'Y'):
                        registration()
                        break
                    else:
                        break
                elif chk == 1:
                    break
                elif chk == 3:
                    inval = input("ENTER 1 FOR FORGOT PASSWORD: ")
                    if inval == '1':
                        forgotPassword(email)
                        break
        
    except FileNotFoundError:
        fname = "login_creds.csv"
        with open(fname, 'a') as f:
            csvWriter = csv.writer(f)
            csvWriter.writerow(["Email", "Password"])
            print("Try Again")

# --------------------------------------------------------------------------------

def forgotPassword(email):
  chk = 0
  fname = "login_creds.csv"
  with open(fname, 'r') as f:
    csvReader = csv.reader(f)
    
    for x in csvReader:
      if email == x[0]:
        print("Email: ", email)
        print("Password: ", x[1])
        return
    if chk == 0:
      print("Login credentials do NOT Exists")

#---------------------------------------------------------------------------------      

# ------------MAIN BLOCK------------
while True:
    print("\nWELCOME \nENTER 1 TO LOGIN \nENTER 2 TO SIGN UP \\ REGISTER \nENTER 3 TO FORGOT PASSWORD \nENTER 4 TO EXIT")
    userIn = input("ENTER YOUR OPTION: ")

    if userIn == "1":
        login()
    elif userIn == "2":
        registration()
    elif userIn == "3":
        email = input("\nEnter Email ID: ")
        forgotPassword(email)
    else:
        print("EXITING.. \nGOODBYE")
        break
```
