###### import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df1= pd.read_csv("/content/college_1.csv", )
df2= pd.read_csv("/content/college_2.csv")
```

```python
df1['Name'] = df1['Name'].str.upper().str.strip()
df1
```

```python
df2['Name'] = df2['Name'].str.upper().str.strip()
df2
```

```python
bgdf= pd.concat([df1,df2], ignore_index= True)
bgdf
```

```python
bgdf['python_en']= bgdf['python_en'].fillna(0)
bgdf['computational_thinking']= bgdf['computational_thinking'].fillna(0)
bgdf
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
bgdf[bgdf['CodeKata Score'] >= 15000].sort_values(by='CodeKata Score', ascending= False).to_csv("/content/Exceeded_expectations.csv")
```

```python
bgdf[(bgdf['CodeKata Score'] >= 10000) & (df1['CodeKata Score'] < 15000)].sort_values(by='CodeKata Score', ascending= False).to_csv("/content/Reached_expectations.csv")
```

```python
bgdf[(bgdf['CodeKata Score'] >= 7000) & (df1['CodeKata Score'] < 10000)].sort_values(by='CodeKata Score', ascending= False).to_csv("/content/Needs_improvement.csv")
```

```python
bgdf[bgdf['CodeKata Score'] < 7000].sort_values(by='CodeKata Score', ascending= False).to_csv("/content/Unsatisfactory.csv")
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

```python
bgdfperf= bgdf.pivot_table(index="Department", values=['Previous Geekions','CodeKata Score', 'Rising'], aggfunc='mean')
bgdfperf['CodeKata Score'],bgdfperf['Previous Geekions'],bgdfperf['Rising']= round(bgdfperf['CodeKata Score'], 2), round(bgdfperf['Previous Geekions'], 2), round(bgdfperf['Rising'], 2)
bgdfperf
```

```python
plt.figure(figsize = (10,8))

plt.plot(bgdfperf['Previous Geekions'])
plt.plot(bgdfperf['CodeKata Score'], color= 'red')
plt.show()
```

###### Department wise codekata performence (pie chart)

```python
piechart= bgdf.pivot_table(index='Department', aggfunc="sum", values=["CodeKata Score", "Previous Geekions", "Rising"] )
piechart
```

```python
fig= plt.figure(figsize = (20,8))
plt.subplot(1,2,1)
plt.pie(piechart['CodeKata Score'], labels=piechart.index, autopct='%1.1f%%')
plt.subplot(1,2,2)
plt.pie(piechart['Previous Geekions'], labels=piechart.index, autopct='%1.1f%%')

plt.show
```
