####Modelling the data

```python
from sklearn.model_selection import train_test_split
X = df[['CRIM',	'ZN', 'INDUS',	'CHAS',	'NOX',	'RM',	'AGE',	'DIS',	'RAD',	'TAX',	'PTRATIO',	'B',	'LSTAT']]
Y = df['MEDV']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

####SIMPLE LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression #importing all the required function

regressor = LinearRegression()
regressor.fit(X_train, Y_train)


Y_pred = regressor.predict(X_test) # given the features of the test dataset, it will give the final predictions
Y_pred
```

```python
df1 = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
df1.head()
```

```python
from sklearn import metrics 
print('R2- SCORE with Simple Linear Regression:', metrics.r2_score(Y_test,Y_pred))
```

####LASSO REGRESSION

```python
from sklearn.linear_model import Lasso #importing all the required function

Lasso_regressor = Lasso()
Lasso_regressor.fit(X_train, Y_train)


Y_pred_lass = Lasso_regressor.predict(X_test) # given the features of the test dataset, it will give the final predictions
Y_pred_lass
```

```python
df2 = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred_lass})
df2.head()
```

```python
from sklearn import metrics 
print('R2- SCORE with Lasso Regression:', metrics.r2_score(Y_test,Y_pred_lass))
```

####RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge #importing all the required function

Ridge_regressor = Ridge()
Ridge_regressor.fit(X_train, Y_train)


Y_pred_ridg = Ridge_regressor.predict(X_test) # given the features of the test dataset, it will give the final predictions
Y_pred_ridg
```

```python
df3 = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred_ridg})
df3.head()
```

```python
from sklearn import metrics 
print('R2- SCORE with Ridge Regression:', metrics.r2_score(Y_test,Y_pred_ridg))
```
