**Step 1: Importing the libraries**

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
```

**Step 2: Importing dataset**

```python
data = pd.read_csv('/content/Data_Assignment11.csv')
data
```

**Step 3: Handling the missing data**

```python
data['Country'].isnull().sum()
```

```python
data['Age'].isnull().sum()
```

```python
data[data['Age'].isnull()]
```

```python
data['Age'] = data.groupby('Country')['Age'].apply(lambda x:x.fillna(x.mean()))

data
```

```python
data['Salary'].isnull().sum()
```

```python
data[data['Salary'].isnull()]
```

```python
data['Salary'] = data.groupby('Country')['Salary'].apply(lambda x:x.fillna(x.mean()))

data
```

**Step 4: Encoding categorical data**

```python
#Since the categorical feature is a multiclass categorical feature and has no order in it, we cannot do Label Encoding and so will perform One-Hot Encoding
```

**Step 5: Creating a dummy variable**

```python
data = pd.get_dummies(data, columns = ['Country'])
data
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X = data[['Age', 'Salary', 'Country_France', 'Country_Germany', 'Country_Spain'] ].values
Y = data['Purchased'].values
#X
#Y

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state= 0)

#X_train
#Y_train
```

**Step 7: Feature Scaling**

```python
scaler = StandardScaler()

scaler.fit(X_train, Y_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

#X_train
```
